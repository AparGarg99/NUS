## **1. Import tensorflow libraries**
---

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import concatenate
from tensorflow.keras.utils import plot_model
```

## **2. Create the model illustrated in DualTw_model.pdf**
---

```python
def createDualTwModel():
    
    #################################### SHARED BLOCK 1 #######################################
    
    # Define shard Layer
    shared = Conv2D(32,(3,3),padding='same',activation='relu')
    
    
    # Left flow
    Lin = Input(shape=(32,32,3))
    
    Lx = Conv2D(16,(3,3),padding='same',activation='relu')(Lin)
    
    Lx = Conv2D(16,(3,3),padding='same',activation='relu')(Lx)
    
    Lx = MaxPooling2D(pool_size=(2,2))(Lx)
    
    Lx = shared(Lx)
    
    # Right flow
    Rin = Input(shape=(16,16,3))
    
    Rx = Conv2D(16,(3,3),padding='same',activation='relu')(Rin)
    
    Rx = shared(Rx)
    
    
    #################################### SHARED BLOCK 2 #######################################
    
    
    # Define shard Layer
    shared = Conv2D(48,(3,3),padding='same',activation='relu')
    
    
    # Left flow
    Lx = MaxPooling2D(pool_size=(2,2))(Lx)
    
    Lx = shared(Lx)
    
    # Right flow
    Rx = MaxPooling2D(pool_size=(2,2))(Rx)
    
    Rx = shared(Rx)
    
    
    #################################### SHARED BLOCK 3 #######################################
    
    # Define shard Layer
    shared = Conv2D(64,(3,3),padding='same',activation='relu')
    
    
    # Left flow
    Lx = MaxPooling2D(pool_size=(2,2))(Lx)
    
    Lx = shared(Lx)
    
    # Right flow
    Rx = MaxPooling2D(pool_size=(2,2))(Rx)
    
    Rx = shared(Rx)
    
    
    ################################### MULTI-INPUT BLOCK 1 #####################################
    
    # Left flow
    Lx = MaxPooling2D(pool_size=(2,2))(Lx)
    
    
    # Right flow
    Rx = MaxPooling2D(pool_size=(2,2))(Rx)
    
    # Concatenate
    x = concatenate([Lx,Rx],axis=-1) 
    
    x = Flatten()(x) 
    
    x = Dense(128,activation='relu')(x) 
    
    x = Dense(3,activation='softmax')(x)
    
    model = Model(inputs=[Lin,Rin],outputs=x)
    
    model.compile(loss='categorical_crossentropy', 
                  optimizer='rmsprop', 
                  metrics=['accuracy'])
    
    return model


modelDualTw = createDualTwModel()
#modelDualTw.summary()
```

## **3. Plot the model**
---

```python
plot_model(modelDualTw, 
           to_file='DualTw_model.png', 
           show_shapes=True, 
           show_layer_names=False,
           rankdir='TB')
```
